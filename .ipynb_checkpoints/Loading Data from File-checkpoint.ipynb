```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.loadtxt("data-01-test-score.csv",delimiter=",",dtype=np.float32)
```

```python
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
```

```python
# make sure the shape and data is ok 
```

```python
print(x_data.shape)
print("\n")
print(y_data.shape)
```

```python
# make the placeholder for the tensor will be always use
X = tf.placeholder(tf.float32,shape=[None,3])
Y = tf.placeholder(tf.float32,shape=[None,1])

W = tf.Variable(tf.random_normal([3,1]),name ="weight")
b = tf.Variable(tf.random_normal([1]),name = 'baise')

# hypothesis 
hypothesis = tf.matmul(X,W)+b
```

```python
# cost function
cost = tf.reduce_mean(tf.square(hypothesis-Y))
```

```python
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
# set session for the graph
sess = tf.Session()
# initilize the global variable
sess.run(tf.global_variables_initializer())
```

```python
for step in range(20001):
    cost_val,hy_val,_ = sess.run([cost,hypothesis,train],feed_dict={X:x_data
                                ,Y:y_data})
    if step % 10 ==0:
        print(step,"cost: ",cost_val,"\n hypothesis: \n",hy_val)
```

```python
# Ask my score
print("Your score will be ", sess.run(hypothesis, 
           feed_dict={X: [[100, 70, 101]]}))

print("Other scores will be ", sess.run(hypothesis, 
           feed_dict={X: [[60, 70, 110], [90, 100, 80]]}))
```
