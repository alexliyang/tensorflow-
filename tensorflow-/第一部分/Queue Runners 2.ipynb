```python
import tensorflow as tf
```

```python
filename_queue = tf.train.string_input_producer(['data-01-test-score.csv'],shuffle=False,name = 'filename_queue')
```

```python
reader = tf.TextLineReader()
key,value = reader.read(filename_queue)
```

```python
# Default values, in case of empty column,Also specifies the type of the decode result
record_defaults = [[0.], [0.], [0.], [0.]]# 在这里其实我们是知道这里面有四列
xy = tf.decode_csv(value, record_defaults=record_defaults)
```

```python
# collect batches of csv in
train_x_batch,train_y_batch = tf.train.batch([xy[0:-1],xy[-1:]],
                                             batch_size = 10)
# 一个取三列 一个取一列
```

```python
# placeholder for a tensor that will always fed
X = tf.placeholder(tf.float32,shape=[None,3])
Y = tf.placeholder(tf.float32,shape=[None,1])

W = tf.Variable(tf.random_normal([3,1]), name = "weight")
b = tf.Variable(tf.random_normal([1]), name = "bias")
```

```python
# Hypothesis
hypothesis = tf.matmul(X,W)+b
```

```python
# simplified cost/loss Function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
# Minilize the cost
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-6)
train = optimizer.minimize(cost)
```

```python
# launch the graph in a session
sess = tf.Session()
# initializes global variable in the graph
sess.run(tf.global_variables_initializer())
```

```python
# start the populating the filename queue
coord = tf.train.Coordinator()#对进程进行协调
threads = tf.train.start_queue_runners(sess = sess,coord=coord)
```

```python
for step in range(2001):
   x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
   cost_val, hy_val, _ = sess.run(
       [cost, hypothesis, train], 
       feed_dict={X: x_batch, Y: y_batch})
   if step % 10 == 0:
       print(step, "Cost: ", cost_val, 
                   "\nPrediction:\n", hy_val)

coord.request_stop()
coord.join(threads)
```
