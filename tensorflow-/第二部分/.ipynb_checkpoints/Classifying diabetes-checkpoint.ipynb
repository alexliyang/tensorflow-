```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.loadtxt('data-03-diabetes.csv',delimiter=",",dtype=np.float32)
print(xy.shape)
```

```python
x_data = xy[:,0:-1]
y_data = xy[:,[-1]]
```

```python
#placeholder for a tensor that will be always fed
X = tf.placeholder(tf.float32,shape=[None,8])
Y = tf.placeholder(tf.float32,shape = [None,1])
W = tf.Variable(tf.random_normal([8,1]),name = "weight")
b = tf.Variable(tf.random_normal([1]),name = "bias")
```

```python
# Hypothesis using sigmod
hypothesis =tf.sigmoid(tf.matmul(X,W)+b)
```

```python
# cost function
cost = -tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*(1-tf.log(hypothesis)))
```

```python
# optimizer
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
# Accuracy # True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis>0.5,dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted,Y),dtype=tf.float32))# nomilize
print(predicted)
```

```python
# lauch the graph
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    feed = {X:x_data,Y:y_data}
    for step in range(10001):
        sess.run(train,feed_dict=feed)
        if step % 100 == 0:
            print(step,sess.run(cost,feed_dict=feed))
    # Accuracy report
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict=feed)
    print("\nHypothesis: ", h, "\nCorrect (Y): ", c, "\nAccuracy: ", a)
```
