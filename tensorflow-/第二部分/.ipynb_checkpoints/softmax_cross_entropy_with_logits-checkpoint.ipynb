```python
import tensorflow as tf
```

```python
import numpy as np
# Predicting animal type based on various features
xy = np.loadtxt('data-04-zoo.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
print(xy.shape)
```

```python
print(xy)
```

```python
nb_classes = 7  # 0 ~ 6
```

```python
X = tf.placeholder(tf.float32, [None, 16])# axis=1为行
Y = tf.placeholder(tf.int32, [None, 1])  # 0 ~ 6
```

```python
Y_one_hot = tf.one_hot(Y, nb_classes)  # one hot

Y_one_hot = tf.reshape(Y_one_hot, [-1, nb_classes])
```

```python
W = tf.Variable(tf.random_normal([16,nb_classes]),name = "weight")
b = tf.Variable(tf.random_normal([7]),name = "bias")
```

```python
# tf.nn.softmax compute softmaxx activation
# softmaxx = exp(logits) / reduce_sum(exp(logits),dim)
logits = tf.matmul(X,W) + b
hypothesis = tf.nn.softmax(logits)
```

```python
# Cross entropy cost/loss
cost_i = tf.nn.softmax_cross_entropy_with_logits(logits = logits,labels = Y_one_hot)
```

```python
cost = tf.reduce_mean(cost_i)
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
prediction = tf.arg_max(hypothesis,1)
correct_prediction = tf.equal(prediction, tf.argmax(Y_one_hot, 1))
accurracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
# lauch graph
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(2000):
        sess.run(train,feed_dict = {X: x_data, Y: y_data})
        if step % 100 == 0:
            loss,acc = sess.run([cost,accurracy],feed_dict={X:x_data, Y:y_data})
            print("Step: {:5}\tLoss: {:.3f}\tAcc: {:.2%}".format(step, loss, acc))

            
            
```
